```python
import subprocess
import os
from datetime import datetime

DIR = "/home/jack/Desktop/HDD500/MyTube_AI_Flask_App/Project_TheOrigin/RAW_Story_resources/mkvideo/"
VIDEOS = "EXPERIMENT/XXXX1.mp4"
MUSIC = "/mnt/HDD500/collections/Music/Music-Copyright_Free_Music.mp3"
directory_name = "EXPERIMENT/"

# Create a filename with the current date and time
current_datetime = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
filename = f"EXPERIMENT/_file_{current_datetime}.mp4"
final_filename = f"{directory_name}{current_datetime}XX.mp4"

# Command 1: Convert images to video
command1 = [
    "ffmpeg",
    "-hide_banner",
    "-i",
    f"{DIR}%*.jpg",
    "-vf",
    "scale=512:768,setsar=1:1",
    "-c:v",
    "libx264",
    "-r",
    "30",
    "-pix_fmt",
    "yuv420p",
    "-y",
    VIDEOS
]

# Command 2: Apply video filters and set duration
command2 = [
    "ffmpeg",
    "-hide_banner",
    "-i",
    VIDEOS,
    "-filter:v",
    "setpts=15*PTS,minterpolate='fps=25:scd=none:me_mode=bidir:vsbmc=1:search_param=200'",
    "-t",
    "58",
    "-y",
    filename
]

# Command 3: Add audio to video and finalize
random_number = str((int(os.urandom(4).hex(), 16) % 3600) + 7200)
command3 = [
    "ffmpeg",
    "-ss",
    "0",
    "-i",
    filename,
    "-ss",
    random_number,
    "-i",
    MUSIC,
    "-map",
    "0:v",
    "-map",
    "1:a",
    "-c:v",
    "copy",
    "-c:a",
    "aac",
    "-shortest",
    "-y",
    final_filename
]

# Execute commands as subprocesses
subprocess.run(command1)
subprocess.run(command2)
subprocess.run(command3)
```

```python
from moviepy.editor import VideoClip
from moviepy.audio.io.AudioFileClip import AudioFileClip
from PIL import Image
import numpy as np
import subprocess
# Convert the MP3 to WAV using FFmpeg
subprocess.run(['ffmpeg','-hide_banner','-i','/home/jack/Desktop/HDD500/MyTube_AI_Flask_App/static/Storie/9.mp3','/home/jack/Desktop/HDD500/MyTube_AI_Flask_App/static/Storie/input_audio_9.wav'])
image_path = '/home/jack/Desktop/HDD500/MyTube_AI_Flask_App/static/Storie/9.jpg'
image_=Image.open(image_path)
audio_path = '/home/jack/Desktop/HDD500/MyTube_AI_Flask_App/static/Storie/input_audio_9.wav'
image_np = np.array(image_)  # Convert Image to NumPy array
image = VideoClip(lambda t: image_np, duration=AudioFileClip(audio_path).duration)
audio = AudioFileClip(audio_path)

output_file = '/home/jack/Desktop/HDD500/MyTube_AI_Flask_App/static/Storie/input_video_9.mp4'
video = image.set_audio(audio)

video.write_videofile(output_file, codec='libx264', audio_codec='aac',fps=24)
```
